```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic_sales_2021_df = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
print(athletic_sales_2020_df)
```

```python
# Display the 2021 sales DataFrame
print(athletic_sales_2021_df)
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_2020_df["retailer"].unique
```

```python
# Check the 2021 sales data types.
athletic_sales_2021_df["retailer"].unique
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
joined_data_rows = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], axis = "rows", join="inner")
joined_data_rows.head(10)

joined_data_rows.reset_index(drop =True, inplace=True)
joined_data_rows.head()
```

```python
# Check if any values are null.
cleaned_joined_data_rows = joined_data_rows.dropna(how="any")
cleaned_joined_data_rows.count()
```

```python
# Check the data type of each column

df = cleaned_joined_data_rows
print(df.dtypes)
```

```python
# Convert the "invoice_date" to a datetime datatype
cleaned_joined_data_rows["invoice_date"] = pd.to_datetime(
    cleaned_joined_data_rows["invoice_date"],
    utc = True
)
```

```python
# Confirm that the "invoice_date" data type has been changed.
cleaned_joined_data_rows.info()

product_subset = cleaned_joined_data_rows[cleaned_joined_data_rows['product']=='Product A']

print(product_subset)
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

products_sold = cleaned_joined_data_rows[["region", "state", "city"]].value_counts()




   



# Show the top 5 results.
products_sold.head()                   
```

#### Using `pivot_table`

### 3. Determine which Region had the Most Sales

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
country_state_total_avg = pd.pivot_table(cleaned_joined_data_rows, 
                                           columns=['city', 'state','region'],
                                           index=['invoice_date'],
                                           values='total_sales',)
# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.
cleaned_joined_data_rows.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
bins = [0,1, 2, 3, 4, 5, 6,7]

# Sort the resampled pivot table in ascending order on "Total Sales".
group_names = ["sunday", "monday", "tuesday", "wednesday", "thursday", "friday","saturday"]

cleaned_joined_data_rows["total_sales"]= pd.cut(cleaned_joined_data_rows["total_sales"],
                                                bins, labels=group_names,
                                                 include_lowest=True)

cleaned_joined_data_rows
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.

country_state_total_avg["total_sales"]= pd.cut(country_state_total_avg["total_sales"]
                                                bins, labels=group_names,
                                                 include_lowest=True)
# Sort the resampled pivot table in ascending order on "Total Sales".
country_state_total_avg
```
